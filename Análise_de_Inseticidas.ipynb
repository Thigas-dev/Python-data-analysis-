```python
#Importanto bibliotecas
import pandas as pd
import numpy as np
```

```python
#pegando a url
url1 = 'https://abre.ai/gb4j'
```

```python
#Data frame dos dados
df1 = pd.read_csv(url1)
df1
```

```python
#gerando grafico de barras
df1.plot.bar()
```

```python
#Agrupamento dos dados baseados no atributo 'A'
#contando e somando os registro/dados
group1 = df1.groupby(['spray'])["count"].sum()
group1
```

```python
group1.plot.bar(color ='purple', title='Eficácia Inseticidas', xlabel = 'inseticidas', ylabel = 'Qtde de insetos')
```

```python
group1.plot(color = 'purple', title='Eficácia Inseticidas', xlabel = 'inseticidas', ylabel = 'Qtde de insetos')
```

```python
group1.plot.bar(color =['purple','black','gold','silver','red','green'], title='Eficácia Inseticidas', xlabel = 'inseticidas', ylabel = 'Qtde de insetos')
```

```python
group1.plot.barh(color =['purple','black','gold','silver','red','green'], title='Eficácia Inseticidas', xlabel = 'Qtde de insetos', ylabel = 'inseticidas')
```

```python
group1.plot.area(color =['purple','black','gold','silver','red','green'], title='Eficácia Inseticidas', ylabel = 'Qtde de insetos', xlabel = 'inseticidas')
```

```python
group1.plot.pie(colors =['purple','orange','gold','silver','red','green'], title='Eficácia Inseticidas', ylabel='Qted de insetos', legend= True, autopct ='%0.2f%%')
```
