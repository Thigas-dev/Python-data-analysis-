```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly.io as pio
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://bit.ly/dadosCEPI', sep = ';')
df.head()
```

```python
df1 = df.iloc[:,0:11]
df1
```

```python
df_turma = df.groupby('Turma').size()
df_turma
```

# Graficos feito na eletiva

```python
# Gráfico de barras

fig = px.bar(df_turma, title = 'Quantidade de Alunos por Turma',
             color_discrete_sequence=px.colors.qualitative.Dark24,
             template='plotly_dark')

# Plotando o gráfico
fig.show()
```

```python
colors = px.colors.qualitative.swatches()
colors.show()
```

#Grafícos da eletiva


```python
# Gráfico de barras

df_idade = df.iloc[:,0:2]
fig = px.bar(df_idade, x = 'Turma', y = 'Idade', color = 'Idade',
             color_discrete_sequence=px.colors.qualitative.Dark24,
             template='plotly_dark')

# Plotando o gráfico
fig.show()
```

```python
# Gráfico de barras

df_idade = df.loc[:,['Turma','Sexo']]
fig = px.bar(df_idade, x = 'Turma', y = 'Sexo', color = 'Sexo', title = "sexo por turma",
             color_discrete_sequence = ['purple','blue','green','black','black'])

# Plotando o gráfico
fig.show()
```

```python
# Gráfico de barras

df_orientacao = df.loc[:,['Turma','Orientação']]
fig = px.bar(df_orientacao, x = 'Turma', y = 'Orientação', color = 'Orientação',title = 'Orientação por turma',
             color_discrete_sequence = ['purple','blue','orange','black','magenta','green'])

# Plotando o gráfico
fig.show()
```

```python
# Gráfico de barras

df_reli_orien = df.loc[:,['Religião','Orientação']]
fig = px.bar(df_reli_orien, x = 'Religião', y = 'Orientação', color = 'Orientação',title = 'Religião x Orientação',
             color_discrete_sequence = ['purple','blue','orange','black','magenta','green'])

# Plotando o gráfico
fig.show()
```

```python
# Gráfico de barras

df_locomo = df.loc[:,['Meio de transporte','Tempo de locomoção']]
fig = px.bar(df_locomo, x = 'Meio de transporte', y = 'Tempo de locomoção', color = 'Tempo de locomoção',title = 'Locomoçao',
             color_discrete_sequence = ['purple','blue','orange','black','magenta','green'])

# Plotando o gráfico
fig.show()
```
