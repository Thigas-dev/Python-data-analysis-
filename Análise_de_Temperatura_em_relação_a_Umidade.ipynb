```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/Ujeverson/datasets/main/tempo.csv'
dados = pd.read_csv(url, sep = ';')
df= pd.DataFrame(data = dados)
df
```

```python
#Se colocar o nome de alguma coluna no s, as bolinhas ficam do tamanho dos dados inseridos
df.plot.scatter(x='Tempo', y='Temperatura', s =150, c='Purple')
```

```python
#Criando o Gráfico para Tempo e Temperatura
plt.scatter(df['Tempo'],df['Temperatura'],s=150,c=['Orange'])
#Criando o gráfico para Tempo e Umidade
plt.scatter(df['Tempo'],df['Umidade'],s=150,c='Blue')
#Colocando legenda
plt.legend(["Temperatura(ºF)","Umidade"],ncol =1, loc="lower right")
plt.show()
```

```python
#Criando o Gráfico para Tempo e Temperatura
plt.scatter(df['Tempo'],df['Temperatura'],s=150,c=['Orange'])
plt.scatter(df['Tempo'],df['Umidade'],s=150,c='Blue')
plt.legend(["Temperatura(ºF)","Umidade"],bbox_to_anchor=(1.62,1),ncol=2)
plt.show()
```
